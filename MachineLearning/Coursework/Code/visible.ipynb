```python
# 导入 pandas 库
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

# 用于忽略 seaborn 产生的一些警告
import warnings
warnings.filterwarnings("ignore")

# 导入 Python 图形库 seaborn
import seaborn as sns
sns.set(style="white", color_codes=True)
```

```python
# 加载 Iris 数据集, 数据在这里"./DataFolder/iris.csv"
trainSet = pd.read_csv("Training-set.csv")

# 查看 Iris 数据集（默认前五项）
trainSet.head()
```

```python
# 查看每个种类有多少项实例
trainSet["t"].value_counts()
```

```python
# 统计四个维度的数据个数、均值、标准差、最大值、最小值等
trainSet[["a", "b", "c"]].describe()
```

```python
trainSet.loc[trainSet["t"]==0][["a", "b", "c"]].describe()
```

```python
trainSet.loc[trainSet["t"]==1][["a", "b", "c"]].describe()
```

```python
# 使用 seaborn.FacetGrid 来绘制散点图，并对不同的物种进行着色
# add_legend()加上图例
sns.FacetGrid(trainSet.loc[(trainSet["c"]<=3)&(trainSet["c"]>=-3)][["a", "b", "c", "t"]], hue="t", size=8).map(plt.scatter, "a", "b").add_legend()

sns.FacetGrid(trainSet, hue="t", size=8).map(plt.scatter, "a", "c").add_legend()
sns.FacetGrid(trainSet, hue="t", size=8).map(plt.scatter, "b", "c").add_legend()
sns.FacetGrid(trainSet, hue="t", size=8).map(plt.scatter, "a", "c").add_legend()
```

```python
# 将四个维度的数据用不同的颜色绘制在一张图上
trainSet[["a", "b", "c"]].plot()
```

```python
# seaborn.pairplot 在任意两个特征之间展示关系
# seaborn.pairplot 的对角上默认显示的是直方图，可以通过参数 diag_kind 赋值的不同来改变
# trainSet.drop("id", axis=1)是将 trainSet 数据集的 id 那一栏暂时丢弃
sns.pairplot(trainSet.drop("id", axis=1), hue="t", diag_kind="kde", size=3)
```

```python
# 使用 seaborn.FacetGrid 来绘制散点图，并对不同的物种进行着色
# add_legend()加上图例
sns.FacetGrid(trainSet.loc[(trainSet['a']<10)&(trainSet['b']<10)&(trainSet['c']<10)], hue="t", size=6).map(plt.scatter, "a", "b").add_legend()
sns.FacetGrid(trainSet.loc[(trainSet['a']<10)&(trainSet['b']<10)&(trainSet['c']<10)], hue="t", size=6).map(plt.scatter, "a", "c").add_legend()
sns.FacetGrid(trainSet.loc[(trainSet['a']<10)&(trainSet['b']<10)&(trainSet['c']<10)], hue="t", size=6).map(plt.scatter, "b", "c").add_legend()
```

```python
# 加载 Iris 数据集, 数据在这里"./DataFolder/iris.csv"
testSet = pd.read_csv("Testing-set.csv")

# 查看 Iris 数据集（默认前五项）
testSet.head()
```

```python
# 统计四个维度的数据个数、均值、标准差、最大值、最小值等
testSet[["a", "b", "c"]].describe()
```

```python
# 使用 seaborn.FacetGrid 来绘制散点图，并对不同的物种进行着色
# add_legend()加上图例
sns.FacetGrid(testSet.loc[(testSet["c"]<=2)&(testSet["c"]>=-2)][["a", "b", "c", "t"]], size=8).map(plt.scatter, "a", "b").add_legend()

sns.FacetGrid(testSet, size=8).map(plt.scatter, "a", "b").add_legend()
sns.FacetGrid(testSet, size=8).map(plt.scatter, "b", "c").add_legend()
sns.FacetGrid(testSet, size=8).map(plt.scatter, "a", "c").add_legend()
```

```python
# kdeplot是与单变量有关的图，可以显示特征的 kernel density estimate
sns.FacetGrid(trainSet, hue="t", size=4).map(sns.kdeplot, "a").add_legend()
sns.FacetGrid(trainSet, hue="t", size=4).map(sns.kdeplot, "b").add_legend()
sns.FacetGrid(trainSet, hue="t", size=4).map(sns.kdeplot, "c").add_legend()
```
